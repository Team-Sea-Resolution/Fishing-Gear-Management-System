```python
import os

# JSON 파일이 있는 폴더 경로를 지정합니다.
# 경로의 시작 부분에 'r'을 붙이면 문자열을 있는 그대로 인식하여 편리합니다.
folder_path = r'C:\Users\USER\Desktop\ocean_data_develop\final_ocean_current_file\json_file'

try:
    # 해당 폴더에 있는 모든 파일과 폴더의 이름을 리스트로 가져옵니다.
    file_list = os.listdir(folder_path)

    # JSON 파일만 필터링하여 이름을 출력합니다.
    print("JSON 파일 목록:")
    for file_name in file_list:
        if file_name.endswith('.json'):
            print(file_name)

except FileNotFoundError:
    print(f"오류: '{folder_path}' 경로를 찾을 수 없습니다. 경로를 다시 확인해주세요.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
import os
import json # JSON 파일을 다루기 위해 추가

# 1. 경로 설정
input_folder = r'C:\Users\USER\Desktop\ocean_data_develop\final_ocean_current_file\json_file'
output_folder = r'C:\Users\USER\Desktop\ocean_data_develop\final_ocean_current_file\processed_files' # 결과를 저장할 새 폴더

# 2. 결과 저장 폴더가 없으면 자동으로 생성
os.makedirs(output_folder, exist_ok=True)

# 3. 입력 폴더의 모든 파일 목록 가져오기
try:
    for file_name in os.listdir(input_folder):
        # 4. JSON 파일만 대상으로 작업 수행
        if file_name.endswith('.json'):
            # 원본 파일의 전체 경로
            input_file_path = os.path.join(input_folder, file_name)

            # -----------------------------------------------------------
            # 여기에 원본 JSON 파일을 읽고 데이터를 처리하는 코드를 작성하세요.
            # 예시: JSON 파일을 열어서 데이터를 읽어오기
            with open(input_file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            # (데이터를 원하는 대로 가공...)
            processed_data = f"이 데이터는 {file_name} 파일로부터 처리되었습니다." # 예시 결과물
            # -----------------------------------------------------------


            # 5. 새 파일 이름 만들기
            # 원본 파일 이름에서 확장자(.json)를 제거합니다.
            base_name = os.path.splitext(file_name)[0]
            
            # 새로운 파일 이름을 정의합니다. (예: 원본이름 + .csv)
            output_file_name = f"{base_name}.csv" 
            # 또는 output_file_name = f"{base_name}_processed.json" 와 같이 할 수도 있습니다.

            # 저장될 파일의 전체 경로
            output_file_path = os.path.join(output_folder, output_file_name)

            # -----------------------------------------------------------
            # 여기에 처리된 데이터를 새 파일에 저장하는 코드를 작성하세요.
            # 예시: 처리된 결과를 텍스트 파일로 저장
            with open(output_file_path, 'w', encoding='utf-8') as f:
                f.write(processed_data)
            # -----------------------------------------------------------

            print(f"처리 완료: '{file_name}' -> '{output_file_name}'")

except FileNotFoundError:
    print(f"오류: '{input_folder}' 경로를 찾을 수 없습니다.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```
